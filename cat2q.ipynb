```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data=pd.read_csv('/content/Facebook.csv')
X=data[['Volume']]
y=data['Close']
regressor=LinearRegression()
regressor.fit(X, y)
data.plot(x='Volume', y='Close', style='o', label='Data Points')
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Volume')
plt.ylabel('Close')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
data=pd.read_csv('/content/Facebook.csv')
X=pd.DataFrame(data['Volume'])
y=pd.DataFrame(data['Close'])
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
y_test.head(5)
```

```python
from sklearn import metrics
import numpy as np
mae=metrics.mean_absolute_error(y_test, y_pred)
mse=metrics.mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import joblib
import pickle
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X, y)
joblib.dump(model, 'linear_regression_model.pkl')
with open('linear_regression_model.pkl', 'wb') as file:pickle.dump(model, file)
```

```python
import joblib
import numpy as np
import pandas as pd
new_X=np.array([600000, 7000000, 8000000, 900000, 1000000]).reshape(-1, 1)
model=joblib.load('linear_regression_model.pkl')
predictions=model.predict(new_X)
new_data=pd.DataFrame({'X': new_X.flatten(), 'Preicted_Y': predictions.flatten()})
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/Facebook.csv')
counties_name='County Name'
unique_counts=data[counties_name].value_counts()
plt.figure(figsize=(10, 6))
plt.plot(unique_counts.index, unique_counts.values, marker='^', linestyle='dotted', color='red')
plt.title(f'Unique Counts of {counties_name} against social media info')
plt.xlabel(counties_name)
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
